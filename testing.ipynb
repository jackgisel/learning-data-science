```python
import pandas as pd
import yfinance as yf
```

```python
df = yf.download("TSLA", start="2020-01-01")
```

```python
df["Pct"] = df.Close.pct_change()
df.dropna(inplace=True)
```

```python
df.resample("M").Pct.sum().plot(kind="bar")
```

```python
df.resample("M").Volume.sum().plot(kind="bar")
```
